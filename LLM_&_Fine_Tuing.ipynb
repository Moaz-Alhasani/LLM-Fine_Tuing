### **Text Generation**

```python
from transformers import pipeline

text_generate = pipeline("text-generation", model="distilgpt2")

generated_text = text_generate(
    "Today is a rainy day in London", truncation=True, max_length=20, do_sample=False
)

print("Generated text:\n", generated_text[0]['generated_text'])
```

## **Question Answering**

```python
qa_model=pipeline("question-answering")
question="What is My job"
context="I am developing Ai models python"
qa_model(question=question,context=context)
```

## **Tokeniztion**

```python
from transformers import AutoTokenizer
tokenizer=AutoTokenizer.from_pretrained("bert-base-uncased")
tokens=tokenizer.tokenize("Using a Transformer network is simple")
print(tokens)
```

## **Convert tokens to inputs Ids**

```python
inputs_ids=tokenizer.convert_tokens_to_ids(tokens)
print(inputs_ids)
```

## **Encode the text**

```python
encode_inputs=tokenizer("Hi everyone i use encode to understand what is the differnts between encode and inputs_ids & attition_mask")
print(encode_inputs)
```

```python
decode=tokenizer.decode(encode_inputs["input_ids"])
print(decode)
```

```python
from datasets import load_dataset
dataset=load_dataset("imdb")
```

```python
from transformers import AutoTokenizer
tokenizer=AutoTokenizer.from_pretrained("bert-base-uncased")
```

```python
def tokenize(examples):
  return tokenizer(examples['text'],padding='max_length',truncation=True)
```

```python
tokenize_datasets=dataset.map(tokenize,batched=True)
```

```python
from transformers import TrainingArguments
training_args=TrainingArguments(
    output_dir='./result',
    eval_steps=2,
    evaluation_strategy="epoch",
    save_strategy="epoch",
    learning_rate=2e-5,
    per_device_train_batch_size=16,
    per_device_eval_batch_size=16,
)
```

```python
from transformers import AutoModelForSequenceClassification
model=AutoModelForSequenceClassification.from_pretrained("bert-base-uncased",num_labels=2)
```

```python
from transformers import Trainer
trainer=Trainer(
    model=model,
    args=training_args,
    train_dataset=tokenize_datasets['train'],
    eval_dataset=tokenize_datasets['test']
)
```
